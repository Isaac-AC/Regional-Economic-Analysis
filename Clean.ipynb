```python
# PIB
df_PIB = pd.read_excel("PIBE_2.xlsx")
df_PIB.head(10)
```

```python
df_PIB.Concepto[12:39]
```

```python
df_PIB.to_csv('PIB_clean.csv', index=False)
```

```python
# total income by state
df_income = pd.read_excel("ingresos.xls")
df_income
```

```python
# Duplicate for not lose the data
df = df_income

# delete the legend "Primer trimestre del XXXX" and left the year
df['trimestre'] = df['trimestre'].str.split().str[-1]

# Rename the column
df = df.rename(columns={'trimestre': 'Year'})
```

```python
# Group by year
df= df.groupby(['Year']).sum().reset_index()

# Sort the values
df = df.sort_values(by='Year', ascending=False)

df.head(3)
```

```python
# Drop the 2023 year
df = df.drop(index= [18, 17])

df.head(3)
```

```python
df.columns
```

```python
df = df.T.reset_index()
df.head(3)
```

```python
# get the first row and save in variable
header = df.iloc[0]

# slice the data leaving the header row
df = df[1:]


# rename the header row as the new dataframe's header
df = df.rename(columns=header)
```

```python
df = df.rename(columns={'Year': 'Estado'})
```

```python
df.to_csv('income_clean.csv', index=False)
```

```python
df_edu = pd.read_excel("Educacion_06.xlsx")
df_edu.head(10)
```

```python
df = df_edu
```

```python
# Utiliza el método str.extract para extraer el año
df['Year'] = df['Ciclo escolar'].str.extract('(\d{4})', expand=False)

# Convierte la columna "Año" a tipo de datos numérico
df['Year'] = pd.to_numeric(df['Year'])

df = df.drop('Ciclo escolar', axis=1)

df
```

```python
df.to_csv('edu_clean.csv', index=False)
```
